```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Parameters for the pendulum
g = 9.8  # gravity
l = 1.0  # length of the pendulum
d = 0.5  # damping coefficient
```

```python
# Define the pendulum equations of motion
def pendulum_ode(y, t, l, d, g):
    theta, theta_dot = y
    dydt = [theta_dot, -(g/l) * np.sin(theta) - d * theta_dot]
    return dydt
```

```python
# Initial conditions
theta0 = [0.75, 0.0]  # Initial angle and angular velocity
```

```python
# Time points at which to compute the solution
t = np.linspace(0, 10, 250)
```

```python
# Solve the differential equation
solution = odeint(pendulum_ode, theta0, t, args=(l, d, g))
```

```python
# Extract theta and theta_dot (x1 and x2)
x1 = solution[:, 0]  # angle (theta)
x2 = solution[:, 1]  # angular velocity (theta_dot)
```

```python
# Plot the solution similar to Figure 1.2
plt.plot(t, x1, label='x1 (theta)')
plt.plot(t, x2, label='x2 (theta_dot)')
plt.xlabel('t')
plt.ylabel('x1, x2')
plt.title('Trajectory of the pendulum')
plt.legend()
plt.grid(True)
plt.ylim([-2, 2])
plt.xlim([0, 10])
plt.show()
```
